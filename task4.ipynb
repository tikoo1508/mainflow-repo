```python
from Tkinter import *
from PIL import ImageTK,Image

#function to initialize the calculator
def cal():
    rt=TK()
    rt.title("Basic calculator")
    
    rt.iconbitmap("C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\Calc-icon.ico") # change the path of icon file
    rt.geometry("300x430")

    bg= ImageTK.PhotoImage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\300x430_background.png")
    bg_label = Label(rt, image= bg)
    bg_label.place(x=0,y=0) # places  bg image on top left corner

    #entry widget
    e = Entry (rt, width=35, borderwidth=10 , font = ("times", 10 ))
    e.grid(row=0,column=0,pady=20 , padx= 30 , columnspan= 3)

    #load images
    images={
        "1":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\1.png")
        "2":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\2.png")
        "3":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\3.png")
        "4":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\4.png")
        "5":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\5.png")
        "6":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\6.png")
        "7":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\7.png")
        "8":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\8.png")
        "9":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\9.png")
        "0":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\10.png")   
        "+":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\11.png")
        "-":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\12.png")
        "*":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\13.png")
        "/":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\14.png")
        "=":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\15.png")
        "C":ImageTK.photoimage(file="C:\Users\tikoo\OneDrive\Desktop\vscode\mainflow\task 4\Calculator Photo\16.png")    
}
    
def btn(number):
        current = e.get()
        e.delete(0,END)
        e.insert(0,str(current) + str(number))
        
def btn_add():
        first_num = e.get()
        global f_num
        global maths
        maths = "addition"
        f_num = int(first_number)
        e.delete (0,END)
        
def btn_sub():
        first_num = e.get()
        global f_num
        global maths
        maths = "subtraction"
        f_num = int(first_number)
        e.delete (0,END)
        
def btn_div():
        first_num = e.get()
        global f_num
        global maths
        maths = "division"
        f_num = int(first_number)
        e.delete (0,END)
        
def btn_mult():
        first_num = e.get()
        global f_num
        global maths
        maths = "multiplication"
        f_num = int(first_number)
        e.delete (0,END)
        
def btn_equal():
        second_number = e.get()
        e.delete(0,END)
        
        if maths==addition:
            e.insert(0,int(f_num)+int(second_number))
        elif maths==subtraction:
            e.insert(0,int(f_num)-int(second_number))
        elif maths == division:
            e.insert(0,int(f_num)/int(second_number))
        elif maths== multiplication:
            e.insert(0,int(f_num)*int(second_number))
            
def btn_clear():
    e.delete(0,END)

buttons = {
    "1":(1,0,lambda : button_click(1)),
    "2":(1,1,lambda : btn(2)),
    "3":(1,2,lambda : btn(3)),
    "4":(2,0,lambda : btn(4)),
    "5":(2,1,lambda : btn(5)),
    "6":(2,2,lambda : btn(6)),
    "7":(3,0,lambda : btn(7)),
    "8":(3,1,lambda : btn(8)),
    "9":(3,2,lambda : btn(9)),
    "0":(4,1,lambda : btn(0)),
    "+":(4,2, btn_add),
    "-":(4,0,btn_sub),
    "/":(5,0,btn_div),
    "*":(5,1,btn_mult),
    "=":(5,2,btn_equal),
    "C":(6,1,btn_clear),
}

for btn , (row,col,cmd) in buttons.items():
    button (rt,border="3", image=images[btn], command =cmd.grid(row=row , column=col))
    
rt.images = images
rt.mainloop()

cal()
```
